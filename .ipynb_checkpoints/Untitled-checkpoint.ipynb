```python
import keras
```

```python
file = open("SMSSpamCollection", "r")
```

```python
s=file.read()
```

```python
print(s)
```

```python
p=s.split("\n")
```

```python
for i in range(len(p)):
    t=p[i].split("\t")
    p[i]=t
```

```python
all_labels = []
all_text = []
for i in range(len(p)):
    try:
        all_text.append(p[i][1])
        all_labels.append(p[i][0])
    except:
        print("NONONO")
```

```python
all_labels
```

```python
all_text
```

```python
for i in range(len(all_text)):
    pt=""
    for j in range(len(all_text[i])):
        if(str(all_text[i][j]).isalpha() or all_text[i][j]==" "):
            pt=pt+str(all_text[i][j]).lower()
    all_text[i]=pt
```

```python
all_text
```

```python
from nltk.corpus import stopwords
all_text2 = []
for i in range(len(all_text)):
    x = all_text[i].split(' ')
    y = [i for i in x if i not in stopwords.words('english')]
    all_text2.append(' '.join(y))
all_text2
```

```python
from keras.preprocessing.text import Tokenizer
tk = Tokenizer(num_words=5000)
tk.fit_on_texts(all_text)
word_index = tk.word_index
seq = tk.texts_to_sequences(all_text)
```

```python
max_len = 0
for i in range(len(seq)):
    if(len(seq[i])>max_len):
        max_len=len(seq[i])
```

```python
seq
```

```python
from keras.preprocessing.sequence import pad_sequences
seq2 = pad_sequences(seq, maxlen=max_len)
```

```python
seq2
```

```python
"""
spam=1
ham=0
"""
rest=[]
for i in range(len(all_labels)):
    if(all_labels[i]=="spam"):
        rest.append(1)
    else:
        rest.append(0)
```

```python
print(len(seq2), len(rest))
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(seq2, rest, test_size = 0.10, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
right_count = 0
for i in range(len(y_pred)):
    if(y_pred[i]==y_test[i]):
        right_count = right_count + 1
print(right_count)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
print(len(X_train), len(y_train))
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=171))
model.add(Dense(128))
model.add(Dense(512))
model.add(Dense(512))
model.add(Dense(128))
model.add(Dense(64))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=64)
```

```python
y_pred2 = [i.round() for i in model.predict(X_test)]
```

```python
right_count = 0
for i in range(len(y_pred2)):
    if(y_pred2[i]==y_test[i]):
        right_count = right_count + 1
print(right_count)
```
