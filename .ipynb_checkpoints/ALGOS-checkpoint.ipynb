```python
file = open("SMSSpamCollection", "r")
```

```python
s=file.read()
```

```python
print(s[0:1000])
```

```python
p=s.split("\n")
```

```python
for i in range(len(p)):
    t=p[i].split("\t")
    p[i]=t
```

```python
all_labels = []
all_text = []
for i in range(len(p)):
    try:
        all_text.append(p[i][1])
        all_labels.append(p[i][0])
    except:
        print("NONONO")
```

```python
all_labels[0:10]
```

```python
all_text[0:10]
```

```python
#Converting all chars to lower case and removing numbers and special characters
for i in range(len(all_text)):
    pt=""
    for j in range(len(all_text[i])):
        if(str(all_text[i][j]).isalpha() or all_text[i][j]==" "):
            pt=pt+str(all_text[i][j]).lower()
    all_text[i]=pt
```

```python
all_text[0:10]
```

```python
#Removing stopwords from the sentences
from nltk.corpus import stopwords
all_text2 = []
for i in range(len(all_text)):
    x = all_text[i].split(' ')
    y = [i for i in x if i not in stopwords.words('english')]
    all_text2.append(' '.join(y))
all_text2[0:10]
```

```python
#Tokenization - max_tokens = 5000
from keras.preprocessing.text import Tokenizer
tk = Tokenizer(num_words=5000)
tk.fit_on_texts(all_text)
word_index = tk.word_index
seq = tk.texts_to_sequences(all_text)
```

```python
max_len = 0
for i in range(len(seq)):
    if(len(seq[i])>max_len):
        max_len=len(seq[i])
print("max length of string: "+str(max_len))
```

```python
seq[0:1]
```

```python
#Padding the arrays
# This step is done because all the rows do NOT have same length of characters.
from keras.preprocessing.sequence import pad_sequences
seq2 = pad_sequences(seq, maxlen=max_len)
```

```python
seq2[0:1]
```

```python
"""
spam=1
ham=0
Label definition
"""
rest=[]
for i in range(len(all_labels)):
    if(all_labels[i]=="spam"):
        rest.append(1)
    else:
        rest.append(0)
```

```python
print(len(seq2), len(rest))
#seq2 = text, rest = labels
```

```python
"""
Here we do train-test split and also feature scaling of the dataset. 
The scaling is bi-polar as the values range from -1 to 1
"""

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(seq2, rest, test_size = 0.10, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
"""
Bayes Algorithm
Probability based
"""
from sklearn.naive_bayes import GaussianNB
c_bayes = GaussianNB()
c_bayes.fit(X_train, y_train)
y_pred_c_bayes = c_bayes.predict(X_test)
right_count = 0
for i in range(len(y_pred_c_bayes)):
    if(y_pred_c_bayes[i]==y_test[i]):
        right_count = right_count + 1
print("accuracy = "+str(right_count/(len(y_pred_c_bayes))*100))
```

```python
"""
Logistic regression algorithm for classification
"""
from sklearn.linear_model import LogisticRegression
c_log = LogisticRegression(random_state = 0)
c_log.fit(X_train, y_train)
y_pred_log = c_log.predict(X_test)
right_count = 0
for i in range(len(y_pred_log)):
    if(y_pred_log[i]==y_test[i]):
        right_count = right_count + 1
print("accuracy = "+str(right_count/(len(y_pred_log))*100))
```

```python
"""
Decision Tree Classifier
"""
from sklearn.tree import DecisionTreeClassifier
c_tree = DecisionTreeClassifier(random_state = 0)
c_tree.fit(X_train, y_train)
y_pred_tree = c_tree.predict(X_test)
right_count = 0
for i in range(len(y_pred_tree)):
    if(y_pred_tree[i]==y_test[i]):
        right_count = right_count + 1
print("accuracy = "+str(right_count/(len(y_pred_tree))*100))
```

```python
"""
Random Forest algorithm for classification
Decision criteria = entropy
"""
from sklearn.ensemble import RandomForestClassifier
c = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
c.fit(X_train, y_train)
y_pred_rf = c.predict(X_test)
right_count = 0
for i in range(len(y_pred_rf)):
    if(y_pred_rf[i]==y_test[i]):
        right_count = right_count + 1
print("accuracy = "+str(right_count/(len(y_pred_rf))*100))
```

# Here is deep learning part with only dense networks.
## Its simple and easy

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=171))
model.add(Dense(128))
model.add(Dense(512))
model.add(Dense(128))
model.add(Dense(64))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=64)
```

```python
y_pred2 = [i.round() for i in model.predict(X_test)]
```

```python
right_count = 0
for i in range(len(y_pred2)):
    if(y_pred2[i]==y_test[i]):
        right_count = right_count + 1
print("accuracy = "+str(right_count/len(y_pred2) * 100))
```
